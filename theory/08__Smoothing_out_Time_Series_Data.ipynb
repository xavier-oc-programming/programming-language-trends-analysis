## **8)—Smoothing out Time-Series Data**

### **Objective**

Reduce noise in time-series data to make underlying trends clearer by applying a **rolling mean**.

---

### **Why Smoothing Is Needed**

Time-series data often contains short-term fluctuations (spikes and dips) that can obscure long-term trends.
Smoothing helps by averaging values over a moving window of time.

* Reveals overall direction and momentum
* Makes comparisons between languages easier
* Reduces visual clutter in multi-line charts

---

### **Key Concepts**

* **Rolling Window**: A fixed number of consecutive observations
* **Rolling Mean**: The average computed within each rolling window
* The window moves forward one observation at a time

Pandas provides two built-in methods that work together:

* `.rolling()`
* `.mean()`

---

### **Creating a Rolling Mean DataFrame**

Example: use a **6-month window**.

```python
# The window is the number of observations being averaged
roll_df = reshaped_df.rolling(window=6).mean()
```

#### **What this does**

* Computes a rolling average for each language
* Produces a new DataFrame (`roll_df`)
* The first few rows will be `NaN` because there are not enough prior values

---

### **Plotting the Smoothed Data**

```python
plt.figure(figsize=(16, 10))
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Number of Posts', fontsize=14)
plt.ylim(0, 35000)

# Plot the smoothed data instead of the original
for column in roll_df.columns:
    plt.plot(
        roll_df.index,
        roll_df[column],
        linewidth=3,
        label=roll_df[column].name
    )

plt.legend(fontsize=16)
```

---

### **What Changed Compared to the Original Chart**

* Lines are smoother and less jagged
* Peaks and declines are easier to interpret
* Long-term trends stand out clearly
* Short-term noise is reduced

---

### **Choosing the Window Size**

The `window` parameter controls how smooth the result is:

* `window=3`

  * Less smoothing
  * More responsive to recent changes

* `window=6`

  * Balanced smoothing
  * Good for monthly data

* `window=12`

  * Heavy smoothing
  * Highlights long-term trends only

Try different values and observe how the chart changes.

---

### **Key Takeaways**

* Rolling means are ideal for noisy time-series data
* Larger windows produce smoother curves
* Smaller windows preserve short-term variation
* Smoothing improves readability and interpretation

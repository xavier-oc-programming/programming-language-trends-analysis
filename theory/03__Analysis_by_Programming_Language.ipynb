## **3)—Solution: Analysis by Programming Language**

### **Objective**

Analyse Stack Overflow post data by programming language to:

* Calculate the **total number of posts per language**
* Determine which languages have been active for the **most and fewest months**

---

### **Key Concept: `groupby()`**

To analyse data by category, we use Pandas’ `.groupby()` method.

* `TAG` is our **categorical column**
* Each unique `TAG` represents a programming language
* Aggregation functions (such as `.sum()` and `.count()`) operate per group

---

### **Total Number of Posts per Language**

To calculate how many posts each programming language has accumulated over time:

```python
df.groupby('TAG').sum()
```

#### **What this does**

* Groups rows by programming language
* Sums the `POSTS` column for each group
* Reveals total Stack Overflow activity per language

#### **Interpretation**

* Languages like **JavaScript, Java, Python, and C#** dominate in total post count
* Older, widely used languages naturally accumulate higher totals

---

### **Counting Months of Activity per Language**

To find how many months each language appears in the dataset:

```python
df.groupby('TAG').count()
```

#### **What this does**

* Counts non-null values per column for each language
* Each count corresponds to a **month with recorded posts**

#### **Why this matters**

* Some languages started later (e.g. Swift)
* Others may have gaps or shorter lifespans in the dataset
* Total post counts must be interpreted in context of activity duration

---

### **Key Observations**

* Most languages appear for **144–145 months**
* **Newer languages** (e.g. Swift) have fewer months of data
* A high total post count may result from:

  * Long lifespan
  * High monthly popularity
  * Or both

---

### **Important Insight**

You should **never compare total post counts alone** without considering:

* Number of months present
* When a language was introduced
* Whether growth is accelerating or declining

This motivates **time-series analysis**, which follows next.

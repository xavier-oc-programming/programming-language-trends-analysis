## **6)—Data Visualisation with Matplotlib**

### **Objective**

Create and style line charts using Matplotlib to visualise programming language popularity over time.

---

### **Introducing Matplotlib**

Matplotlib is a core Python library for creating charts and graphs.
It integrates seamlessly with Pandas and is ideal for time-series visualisation.

---

### **Importing Matplotlib**

Always import Matplotlib at the top of your notebook.

```python
import matplotlib.pyplot as plt
```

---

### **Creating a Basic Line Chart**

You can create a line chart with a **single line of code** using `.plot()`.

Example: plot the popularity of **Java** over time.

```python
plt.plot(reshaped_df.index, reshaped_df.java)
```

Equivalent square-bracket notation:

```python
plt.plot(reshaped_df.index, reshaped_df['java'])
```

#### **Explanation**

* x-axis: dates (`reshaped_df.index`)
* y-axis: number of posts for Java
* Matplotlib automatically renders the chart

---

### **Resizing the Figure**

Larger charts are easier to read.

```python
plt.figure(figsize=(16, 10))
plt.plot(reshaped_df.index, reshaped_df.java)
```

* `figsize=(width, height)` is measured in inches

---

### **Styling the Axes**

To keep the chart readable when increasing size, also increase tick font sizes.

```python
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
```

---

### **Adding Axis Labels and Limits**

Label your axes clearly and define sensible bounds.

```python
plt.xlabel('Date', fontsize=14)
plt.ylabel('Number of Posts', fontsize=14)
plt.ylim(0, 35000)
```

#### **Why `ylim(0, ...)`**

* Post counts cannot be negative
* Prevents misleading visual scaling

---

### **Complete Styled Example (Single Language)**

```python
plt.figure(figsize=(16, 10))
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Number of Posts', fontsize=14)
plt.ylim(0, 35000)
plt.plot(reshaped_df.index, reshaped_df.java)
```

---

## **Challenge: Plot Multiple Languages**

### **Goal**

Plot **Java** and **Python** on the same chart to compare trends.

---

### **Solution**

```python
plt.figure(figsize=(16, 10))
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Number of Posts', fontsize=14)
plt.ylim(0, 35000)

plt.plot(reshaped_df.index, reshaped_df.java)
plt.plot(reshaped_df.index, reshaped_df.python)
```

---

### **Key Observations**

* Java peaked earlier and then declined
* Python shows sustained growth and overtakes Java
* Plotting multiple lines reveals comparative trends instantly

---

### **Key Takeaways**

* `.plot()` creates line charts quickly
* `figure()`, `xlabel()`, `ylabel()`, and `ylim()` improve clarity
* Multiple calls to `.plot()` draw multiple lines on the same chart
* Time-series data becomes meaningful when visualised

---

### **End Result**

By the end of this section, you can:

* Create clean time-series plots
* Style charts for readability
* Compare multiple programming languages visually

### **Reading the CSV File**

We load the dataset using `read_csv()` and explicitly define the column names.

```python
df = pd.read_csv(
    'QueryResults.csv',
    names=['DATE', 'TAG', 'POSTS'],
    header=0
)
```

#### **Explanation**

* `names=[...]` overrides the column names
* `header=0` tells Pandas to treat the first row as data, not headers
* This guarantees consistent column naming regardless of the CSV structure

## **7)—Multi-Line Charts with Matplotlib**

### **Objective**

Plot multiple programming languages on the same chart and make the visualization readable by:

* Automating plotting with a loop
* Adding labels and a legend
* Improving line visibility

---

### **Plotting Two Lines on the Same Chart**

The simplest way to draw multiple lines is to call `.plot()` multiple times.

```python
plt.figure(figsize=(16, 10))
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Number of Posts', fontsize=14)
plt.ylim(0, 35000)

plt.plot(reshaped_df.index, reshaped_df.java)
plt.plot(reshaped_df.index, reshaped_df.python)
```

#### **Key Idea**

* Each call to `.plot()` adds a new line to the same figure
* This works, but does not scale well when plotting many languages

---

### **Plotting All Languages with a Loop**

Instead of writing one `.plot()` per language, iterate over the columns.

```python
for column in reshaped_df.columns:
    plt.plot(reshaped_df.index, reshaped_df[column])
```

#### **Why This Works**

* Each column corresponds to a programming language
* The loop automatically plots every language in the dataset
* The chart updates once per iteration

---

### **Improving Readability: Line Width and Labels**

Without labels, it is impossible to tell which line belongs to which language.
We fix this by:

* Adding a label to each line
* Making lines thicker

```python
plt.figure(figsize=(16, 10))
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Number of Posts', fontsize=14)
plt.ylim(0, 35000)

for column in reshaped_df.columns:
    plt.plot(
        reshaped_df.index,
        reshaped_df[column],
        linewidth=3,
        label=reshaped_df[column].name
    )
```

---

### **Adding a Legend**

The legend maps colors to programming languages.

```python
plt.legend(fontsize=16)
```

#### **Result**

* Each line is clearly labeled
* Trends can be compared across all languages
* The chart becomes interpretable instead of cluttered

---

### **Key Observations**

* Python shows sustained growth and overtakes other languages
* Older languages peak earlier and flatten or decline
* Multi-line charts reveal relative trends that single-line charts cannot

---

### **Key Takeaways**

* Multiple `.plot()` calls draw multiple lines
* Loops make plotting scalable
* `label` and `legend()` are essential for clarity
* Line width improves visibility when many lines overlap

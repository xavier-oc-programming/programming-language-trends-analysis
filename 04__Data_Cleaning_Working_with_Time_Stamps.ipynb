## **4)—Data Cleaning: Working with Time Stamps**

### **Objective**

Convert the `DATE` column from plain strings into proper datetime objects so the data can be analysed and visualised as a time series.

---

### **Selecting an Individual Cell**

To inspect individual values inside a column, we can access a specific cell using **indexing**.

#### **Square Bracket Notation**

```python
df['DATE'][1]
```

#### **Dot Notation (when column names have no spaces)**

```python
df.DATE[1]
```

#### **Notes**

* Square bracket notation is **more flexible**
* Dot notation enables **autocomplete** in VS Code
* Both return the same value

---

### **Inspecting the Data Type**

Although the values look like dates, they are actually stored as **strings**.

```python
type(df['DATE'][1])
```

#### **Result**

```text
str
```

---

### **Why This Is a Problem**

Working with dates as strings has several drawbacks:

* Time information (`00:00:00`) is always displayed
* No built-in support for:

  * Sorting by date
  * Resampling
  * Time-based grouping
* Plotting time-series becomes unreliable

To fix this, we need **datetime objects**, not strings.

---

### **Converting a Single Value to Datetime**

Pandas provides the `to_datetime()` function to convert strings into timestamps.

```python
pd.to_datetime(df.DATE[1])
```

To confirm the conversion:

```python
type(pd.to_datetime(df.DATE[1]))
```

#### **Result**

```text
pandas._libs.tslibs.timestamps.Timestamp
```

This confirms the value is now a proper datetime object.

---

### **Converting the Entire DATE Column**

Instead of converting individual values, we convert the **entire column at once**.

```python
df.DATE = pd.to_datetime(df.DATE)
```

After conversion, inspect the result:

```python
df.head()
```

#### **What Changes**

* Dates are now shown in a clean `YYYY-MM-DD` format
* The column is ready for:

  * Time-series analysis
  * Pivoting
  * Plotting with Matplotlib

---

### **Key Takeaway**

Always ensure that date columns are stored as **datetime objects**, not strings.
This is essential for any serious data analysis involving time.

---

### **End Result**

By the end of this section, you now have:

* A clean datetime column
* Proper time-aware data
* A dataset ready for reshaping and visualisation

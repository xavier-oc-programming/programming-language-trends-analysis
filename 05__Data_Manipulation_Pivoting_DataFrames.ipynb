## **5)—Data Manipulation: Pivoting DataFrames**

### **Objective**

Reshape the dataset so that:

* Each **row** represents a date
* Each **column** represents a programming language
* Each cell contains the **number of Stack Overflow posts** for that language and month

This structure is required for time-series visualisation.

---

### **The `pivot()` Method**

Sometimes data is stored in a “long” format (many rows, few columns), but for analysis and plotting we need a “wide” format.

The `pivot()` method allows us to:

* Choose which column becomes the **index**
* Choose which column becomes the **new columns**
* Choose which values populate the cells

---

### **Conceptual Example**

Original structure:

* One column for categories (e.g. Actor)
* One column for values (e.g. Power)

After pivoting:

* Each category becomes its **own column**
* Rows represent another variable (e.g. Age)
* Missing combinations become `NaN`

This exact idea applies to programming languages and dates.

---

### **Pivoting the Stack Overflow Dataset**

To reshape our existing DataFrame:

```python
reshaped_df = df.pivot(
    index='DATE',
    columns='TAG',
    values='POSTS'
)
```

#### **What this does**

* `DATE` becomes the row index
* Each programming language (`TAG`) becomes a column
* Each cell contains the number of posts for that language in that month

---

### **Inspecting the Reshaped Data**

Check the dimensions:

```python
reshaped_df.shape
```

#### **Result**

* **145 rows**
* **14 columns**

This corresponds to:

* 145 months of data
* 14 programming languages

---

### **Inspecting Columns and Data Preview**

```python
reshaped_df.columns
```

This confirms that:

* Each programming language is now a separate column

Preview the data:

```python
reshaped_df.head()
```

You will notice `NaN` values where:

* A language had no posts in a given month
* The language did not yet exist (e.g. Swift in early years)

---

### **Counting Entries per Language**

```python
reshaped_df.count()
```

#### **Important Detail**

* `.count()` ignores `NaN` values
* Languages with fewer months of data appear with smaller counts

This confirms:

* Older languages have more complete timelines
* Newer languages naturally have fewer entries

---

### **Dealing with NaN Values**

For plotting and analysis, we **do not want missing values**.
Instead of dropping rows, we replace `NaN` with `0`.

```python
reshaped_df.fillna(0, inplace=True)
```

#### **Why `inplace=True`**

* Modifies the existing DataFrame directly
* Avoids reassigning the variable

Equivalent alternative:

```python
reshaped_df = reshaped_df.fillna(0)
```

---

### **Verifying That No NaN Values Remain**

```python
reshaped_df.isna().values.any()
```

#### **Result**

```text
False
```

This confirms:

* All missing values have been successfully replaced
* The dataset is now safe for plotting

---

### **Key Takeaways**

* `pivot()` reshapes long-format data into wide-format
* Missing category combinations become `NaN`
* `fillna(0)` is essential before time-series plotting
* Clean, rectangular data is required for Matplotlib

---

### **End Result**

By the end of this section, you now have:

* One column per programming language
* One row per month
* No missing values
* A dataset ready for multi-line time-series charts

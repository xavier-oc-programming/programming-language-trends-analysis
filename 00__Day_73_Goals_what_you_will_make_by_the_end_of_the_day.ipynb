### **Objective**

Analyse the **popularity of different programming languages over time** using real-world data and visualise trends clearly with Matplotlib.

---

### **Context**

Programming languages constantly rise and fall in popularity. Some become dominant, others fade, and a few remain relevant for decades.

To answer the question *“Which programming language is the most popular?”*, this lesson uses **Stack Overflow data**, where:

* Each post is tagged with one or more programming languages
* The number of posts per tag reflects popularity
* Time-based data allows trend analysis across years

The language with the **highest number of tagged posts** is considered the most popular.

---

### **Data Source**

* **Stack Overflow**
* Tags represent programming languages
* Popularity is measured by **post count over time**
